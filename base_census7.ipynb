```python
from sklearn.preprocessing import LabelEncoder
```

```python
xbase = basecensu.iloc[:, 0:14].values
```

```python
ybase = basecensu.iloc[:,14].values
```

```python
labelencoder = LabelEncoder()
xbase[:, 1] = labelencoder.fit_transform(xbase[:, 1])
xbase[:, 3] = labelencoder.fit_transform(xbase[:, 3])
xbase[:, 5] = labelencoder.fit_transform(xbase[:, 5])
xbase[:, 6] = labelencoder.fit_transform(xbase[:, 6])
xbase[:, 7] = labelencoder.fit_transform(xbase[:, 7])
xbase[:, 8] = labelencoder.fit_transform(xbase[:, 8])
xbase[:, 9] = labelencoder.fit_transform(xbase[:, 9])
xbase[:, 13] = labelencoder.fit_transform(xbase[:, 13])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
onehot = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,3,5,6,7,8,9,13])],remainder = 'passthrough')
```

```python
xbase = onehot.fit_transform(xbase).toarray()
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
xbase = standardscaler.fit_transform(xbase)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtreino, xtest, ytreino, ytest = train_test_split(xbase,ybase, test_size=0.25, random_state = 0)
```

```python
xtreino.shape, ytreino.shape
```

```python
xtest.shape, ytest.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
redeneural_censu = MLPClassifier(max_iter=1000, verbose=True, tol = 0.000010,hidden_layer_sizes=(55,55))
```

```python
redeneural_censu.fit(xtreino, ytreino)
```

```python
previsoes = redeneural_censu.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(previsoes,ytest)
```
