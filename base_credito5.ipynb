```python
x_credit = basecredito.iloc[:, 1:4].values
x_credit
```

```python
y_credit = basecredito.iloc[:, 4].values
y_credit
```

```python
from sklearn.preprocessing import StandardScaler
scaler_credit = StandardScaler()
x_credit = scaler_credit.fit_transform(x_credit)
x_credit
```

```python
from sklearn.model_selection import train_test_split
```

```python
xcredito_treino, xcredito_test, ycredito_treino, ycredito_test = train_test_split(x_credit, y_credit, test_size = 0.25, random_state = 0)
```

```python
xcredito_treino.shape, ycredito_treino.shape
```

```python
from sklearn.svm import SVC
```

```python
svm_credit = SVC(kernel = 'rbf',random_state=1, C=2.0)
svm_credit.fit(xcredito_treino, ycredito_treino)
```

```python
previsoes = svm_credit.predict(xcredito_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(previsoes, ycredito_test)
```

```python
print(classification_report(previsoes,ycredito_test))
```
