```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
basecasas = pd.read_csv('/content/house_prices.csv')
basecasas
```

```python
basecasas.describe()
```

```python
basecasas.isnull().sum()
```

```python
basecasas['date'] = pd.to_datetime(basecasas['date'])
```

```python
correlation_matrix = basecasas.select_dtypes(include=['number']).corr()
print(correlation_matrix)
```

```python
figura = plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, annot = True);
```

```python
xcasa = basecasas.iloc[:,5:6].values
xcasa
```

```python
ycasa = basecasas.iloc[:,2].values
ycasa
```

```python
from sklearn.model_selection import train_test_split
xtreino, xtest, ytreino,ytest = train_test_split(xcasa,ycasa, test_size=0.3, random_state=0)
```

```python
xtreino.shape, ytreino.shape
```

```python
xtest.shape, ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
regressorsimples = LinearRegression()
regressorsimples.fit(xtreino,ytreino)
```

```python
#b0
regressorsimples.intercept_
```

```python
#b1
regressorsimples.coef_
```

```python
regressorsimples.score(xtreino,ytreino)
```

```python
regressorsimples.score(xtest,ytest)
```

```python
previsoes = regressorsimples.predict(xtreino)
previsoes
```

```python
grafico = px.scatter(x = xtreino.ravel(), y = previsoes)
```

```python
grafico.show()
```

```python
grafico1 = px.scatter(x= xtreino.ravel(), y = ytreino)
grafico2 = px.line(x=xtreino.ravel(), y = previsoes)
grafico2.data[0].line.color ='red'
grafico3= go.Figure(data=grafico1.data + grafico2.data)
grafico3
```
