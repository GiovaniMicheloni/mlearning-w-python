ECLAT - Algoritmo utilizado para descobrir conjuntos de itens frequentes em um banco de dados transacional.

```python
!pip install pyECLAT
```

```python
import pandas as pd
```

```python
base_mercado1 = pd.read_csv('mercado.csv', header = None)
base_mercado1
```

```python
from pyECLAT import ECLAT
eclat = ECLAT(data = base_mercado1)
```

```python
eclat.df_bin
```

```python
eclat.uniq_
```

```python
indices, suporte = eclat.fit(min_support=0.3, min_combination=1, max_combination=3)
```
