```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_saude = pd.DataFrame({
    'Idade': [18,23,28,33,38,43,48,53,58,63],
    'Custo': [470,520,630,830,1150,1530,2040,3080,5100,10100]
})
```

```python
df_saude
```

```python
xplano = df_saude.iloc[:,0]
```

```python
xplano
```

```python
yplano = df_saude.iloc[:,1]
yplano
```

```python
np.corrcoef(xplano, yplano)
```

```python
xplano = df_saude.iloc[:,0].values.reshape(-1,1)
xplano.shape
```

```python
 from sklearn.tree import DecisionTreeRegressor
 regressor_arvore_saude = DecisionTreeRegressor()
 regressor_arvore_saude.fit(xplano,yplano)
```

```python
previsoes = regressor_arvore_saude.predict(xplano)
previsoes
```

```python
regressor_arvore_saude.score(xplano,yplano)
```

```python
grafico = px.scatter(x = xplano.ravel(), y = yplano)
grafico.add_scatter(x = xplano.ravel(), y = previsoes, name ='Regressão')
grafico.show()
```

```python
xtest_arvore = np.arange(min(xplano),max(xplano), 0.1)
xtest_arvore
```

```python
xtest_arvore.shape
```

```python
xtest_arvore = xtest_arvore.reshape(-1,1)
xtest_arvore.shape
```

```python
grafico = px.scatter(x = xplano.ravel(), y = yplano)
grafico.add_scatter(x = xtest_arvore.ravel(), y = regressor_arvore_saude.predict(xtest_arvore), name ='Regressão')
grafico.show()
```
