```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_saude = pd.DataFrame({
    'Idade': [18,23,28,33,38,43,48,53,58,63],
    'Custo': [871,1132,1242,1356,1488,1638,2130,2454,3066,4090]
})
```

```python
df_saude
```

```python
xplano = df_saude.iloc[:,0]
xplano
```

```python
yplano = df_saude.iloc[:,1]
yplano
```

```python
np.corrcoef(xplano, yplano)
```

```python
xplano.shape
```

```python
xplano = xplano.reshape(-1,1)
xplano.shape
```

```python
from sklearn.linear_model import LinearRegression
regressorplano = LinearRegression()
regressorplano.fit(xplano,yplano)
```

```python
#b0 intercepta o eixo y
regressorplano.intercept_
```

```python
#b1 inclinaçao da reta
regressorplano.coef_
```

```python
previsoes = regressorplano.predict(xplano)
previsoes  #y=b0+b1*x1 valor=ponto que intercept+inclinação * idade
```

```python
yplano
```

```python
xplano = df_saude['Idade'].to_numpy().reshape(-1, 1)
```

```python
grafico = px.scatter(x = xplano.ravel(), y = yplano)
grafico.add_scatter(x = xplano.ravel(), y = previsoes, name ='Regressão')
grafico.show()
```
